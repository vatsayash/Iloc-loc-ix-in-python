# Selection uding iloc

```python
# Rows and columns selection using iloc in DataFrame
# Rows
data.iloc[0] #selects first row of the data
data.iloc[1] #selects the second row of the data
data.iloc[-1] #selects the last row of the data
#columns
data.iloc[:,0] #selects the first column of the dataset
data.iloc[:,1] #selects the second column onf the dataset
data.iloc[:,-1] # selects the last column of the dataset
```

## Multiple rows and columns selection in using iloc

```python
data.iloc[:5] # selects the first 5 rows of the dataset
data.iloc[:,:2] # selects all rows but only first two columns
data.iloc[[1,2,4],[5,6]] # selects the corresponding rows and columns
data.iloc[0:5,0:8]
```

Note that .iloc returns a Pandas Series when one row is selected, and a Pandas DataFrame when multiple rows are selected, or if any column in full is selected. To counter this, pass a single-valued list if you require DataFrame output.

Pandas loc indexer can be used with two different use cases
1) Selecting rows by label/index
2) selecting by boolean condition


```python
data.set_index("last_name",inplace = True)
data.head()
```

```python
data.loc['Andrade']
```

```python
data.loc[['Andrade','Veness']]
```

```python
# Select rows with index values 'Andrade' and 'Veness', with all columns between 'city' and 'email'
data.loc[['Andrade', 'Veness'], 'city':'email']
```

```python
# Select same rows, with just 'first_name', 'address' and 'city' columns
data.loc['Andrade':'Veness', ['first_name', 'address', 'city']]
```

```python
# Change the index to be based on the 'id' column
data.set_index('id', inplace=True)
# select the row with 'id' = 487
data.loc[717]
```

 Boolean / Logical indexing using .loc

```python
data.loc[data['first_name'] == 'Antonio']
```

```python
data.loc[data['first_name'] == 'Erasmo',['company_name','address','city']]
```

```python
data.loc[data['first_name'] == 'Antonio','email'] #returns a series
```

```python
data.loc[data['first_name'] == 'Antonio',['email']]  # returns a data frame
```

```python
data.loc[data['first_name'] == 'Antonio','city':'email']
data.loc[data['email'].str.endswith('hotmail.com')]
data.loc[data['email'].str.endswith("gmail.com") & (data['first_name'] == 'Antonio')] 
data.loc[(data['id'] > 100) & (data['id'] <= 200), ['postal', 'web']] 
data.loc[data['company_name'].apply(lambda x: len(x.split(' ')) == 4)]
idx = data['company_name'].apply(lambda x: len(x.split(' ')) == 4)
data.loc[idx, ['email', 'first_name', 'company']]
```
